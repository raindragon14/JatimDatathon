# Processing Modal Usaha

## Load the dataset

### Subtask:
Load the `nilai_modal_usaha_atas_izin_usaha_kecil_menurut_kabkota.csv` file into a pandas DataFrame.


**Reasoning**:
The subtask is to load a CSV file into a pandas DataFrame. The provided instructions outline the necessary steps: importing pandas, reading the CSV, and displaying the head of the DataFrame. These can be done in a single code block.



```python
import pandas as pd

df_modal_usaha = pd.read_csv('/content/nilai_modal_usaha_atas_izin_usaha_kecil_menurut_kabkota.csv')
display(df_modal_usaha.head())
```

## Clean 'modal usaha' column

### Subtask:
Convert the 'modal_usaha__triliun_rp_' column to a numeric type, handling the comma as a decimal separator.


**Reasoning**:
The task requires converting the 'modal_usaha__triliun_rp_' column to a numeric type, which involves replacing the comma decimal separator with a period and then casting the column to a numeric data type.



```python
df_modal_usaha['modal_usaha__triliun_rp_'] = df_modal_usaha['modal_usaha__triliun_rp_'].str.replace(',', '.', regex=False)
df_modal_usaha['modal_usaha__triliun_rp_'] = pd.to_numeric(df_modal_usaha['modal_usaha__triliun_rp_'])
display(df_modal_usaha.info())
```

## Extract year

### Subtask:
Extract the year from the 'periode_update' column.


**Reasoning**:
Extract the year from the 'periode_update' column using string slicing and store it in a new 'tahun' column, then display the head of the DataFrame.



```python
df_modal_usaha['tahun'] = df_modal_usaha['periode_update'].astype(str).str[:4]
display(df_modal_usaha.head())
```

## Aggregate data

### Subtask:
Group the DataFrame by 'nama_kabupaten_kota' and 'tahun' and sum the 'modal_usaha__triliun_rp_' to accumulate quarterly data.


**Reasoning**:
Group the DataFrame by 'nama_kabupaten_kota' and 'tahun' and sum the 'modal_usaha__triliun_rp_' to accumulate quarterly data.



```python
df_aggregated = df_modal_usaha.groupby(['nama_kabupaten_kota', 'tahun'])['modal_usaha__triliun_rp_'].sum().reset_index()
display(df_aggregated.head())
```

## Select and rename columns

### Subtask:
Select the aggregated 'nama_kabupaten_kota', the summed 'modal_usaha__triliun_rp_', and 'tahun' columns from `df_aggregated` and rename them to 'kabupaten/kota', 'modal usaha', and 'tahun' respectively.


**Reasoning**:
Select and rename the required columns from the aggregated DataFrame.



```python
df_processed = df_aggregated[['nama_kabupaten_kota', 'modal_usaha__triliun_rp_', 'tahun']].copy()
df_processed.rename(columns={'nama_kabupaten_kota': 'kabupaten/kota', 'modal_usaha__triliun_rp_': 'modal usaha'}, inplace=True)
display(df_processed.head())
```

## Save the new dataset

### Subtask:
Save the processed DataFrame `df_processed` to a new CSV file named "processed_modal_usaha.csv".


**Reasoning**:
Save the processed DataFrame `df_processed` to a CSV file named "processed_modal_usaha.csv" without including the index.



```python
df_processed.to_csv('processed_modal_usaha.csv', index=False)
```

## Provide download link

### Subtask:
Generate a download link for the newly created CSV file, "processed_modal_usaha.csv".


**Reasoning**:
Use the `files.download` function to provide a download link for the created CSV file.



```python
from google.colab import files

files.download('processed_modal_usaha.csv')
```

### Data Analysis Key Findings

*   The original dataset contained quarterly data for modal usaha across different regencies/cities and years.
*   The `modal_usaha__triliun_rp_` column, initially stored as text with commas as decimal separators, was successfully converted to a numeric format (float64) by replacing commas with periods.
*   The year was successfully extracted from the `periode_update` column, allowing for aggregation by year.
*   The data was aggregated by `nama_kabupaten_kota` and `tahun`, summing the `modal_usaha__triliun_rp_` to provide the annual total modal usaha for each regency/city.
*   A new DataFrame was created with the required columns (`kabupaten/kota`, `modal usaha`, and `tahun`) and saved as "processed_modal_usaha.csv".

#Master Dataset 2022

```python
import sqlite3
import pandas as pd

# Create an in-memory SQLite database
conn = sqlite3.connect(':memory:')

try:
    df_dataset1 = pd.read_csv('/content/2022/Garis Kemiskinan Menurut Kabupaten_Kota di Jawa Timur, 2022.csv')
    df_dataset2 = pd.read_csv('/content/2022/Indeks Pembangunan Manusia (Hasil Long Form SP2020) Menurut Kabupaten_Kota, 2022.csv')
    df_dataset3 = pd.read_csv('/content/2022/Persentase Penduduk Miskin Menurut Kabupaten_Kota di Jawa Timur, 2022.csv')
    df_dataset4 = pd.read_csv('/content/2022/Modal_Usaha.csv')
    df_dataset5 = pd.read_csv('/content/2022/persentase_koperasi_aktif.csv')
    df_dataset6 = pd.read_csv('/content/persentase_usaha_mikro_dan_kecil.csv')
except FileNotFoundError:
    print("Make sure your CSV files are in the correct path.")
    # For demonstration, let's keep the sample dataframes if files are not found
    data1 = {'Kabupaten/Kota Se Jawa Timur': ['A', 'B', 'A', 'C'], 'col2': [1, 2, 3, 4]} # Added sample data for Kabupaten/Kota Se Jawa Timur
    df_dataset1 = pd.DataFrame(data1)

    data2 = {'Kabupaten/Kota Se Jawa Timur': ['A', 'B', 'D'], 'col3': ['X', 'Y', 'Z']} # Added sample data for Kabupaten/Kota Se Jawa Timur
    df_dataset2 = pd.DataFrame(data2)


# Load your DataFrames into the database as tables with simpler names
df_dataset1.to_sql('GarisKemiskinan_2022', conn, index=False, if_exists='replace')
df_dataset2.to_sql('IPM_2022', conn, index=False, if_exists='replace')
df_dataset3.to_sql('PersentaseKemiskinan_2022', conn, index=False, if_exists='replace')
df_dataset4.to_sql('Modal_Usaha', conn, index=False, if_exists='replace')
df_dataset5.to_sql('KoperasiAktif', conn, index=False, if_exists='replace')
df_dataset6.to_sql('PersenUsahaKecil', conn, index=False, if_exists='replace')

# Now you can perform SQL queries to recreate or create new datasets
# Example: Create a new dataset by selecting distinct kabupaten/kota and skipping None
query = """
SELECT
    gk."Kabupaten/Kota Se Jawa Timur" AS Nama_Kabupaten_Kota,
    gk."Garis Kemiskinan Menurut Kabupaten/Kota di Jawa Timur (Rupiah/Bulan/Kapita)" AS Garis_Kemiskinan_2022,
    ipm."Indeks Pembangunan Manusia (Hasil Long Form SP2020) Menurut Kabupaten/Kota" AS Indeks_Pembangunan_Manusia,
    pk."Persentase Penduduk Miskin Menurut Kabupaten/Kota di Jawa Timur (Persen)" AS Persen_Penduduk_Miskin,
    mu."modal_usaha" AS Nilai_Modal_Usaha_Kecil,
    ka."jumlah_koperasi_aktif" AS Persen_Koperasi_Aktif,
    pu."jumlah" AS Persen_UMK
FROM
    GarisKemiskinan_2022 AS gk -- Using the simpler table name
LEFT JOIN
    IPM_2022 AS ipm
ON
    gk."Kabupaten/Kota Se Jawa Timur" = ipm."Kabupaten/Kota Se Jawa Timur"
LEFT JOIN
    PersentaseKemiskinan_2022 AS pk
ON
    gk."Kabupaten/Kota Se Jawa Timur" = pk."Kab/Kota"
LEFT JOIN
    Modal_Usaha AS mu
ON
    gk."Kabupaten/Kota Se Jawa Timur" = mu."kabupaten/kota" AND mu.tahun = 2022
LEFT JOIN
    KoperasiAktif AS ka
ON gk."Kabupaten/Kota Se Jawa Timur" = ka."nama_kabupaten_kota" AND ka.kategori = 'PERSENTASE RAT' AND ka.tahun = 2022
LEFT JOIN
    PersenUsahaKecil AS pu
ON gk."Kabupaten/Kota Se Jawa Timur" = pu."nama_kabupaten_kota" AND pu.kategori = 'PERSENTASE USAHA KECIL' AND pu.tahun = 2022;
"""

# Execute the query and fetch the results into a new pandas DataFrame
master_dataset_2022_df = pd.read_sql_query(query, conn)

# Display the new dataset
display(master_dataset_2022_df)

# Close the connection when done
conn.close()
```

```python
from google.colab import files

# Save the master dataset to a CSV file
master_dataset_df.to_csv('master_dataset_2022.csv', index=False)

# Provide a download link for the CSV file
files.download('master_dataset_2022.csv')
```
